### Set up normalization, dataset

```python
# I need to make this a function, since I'll be applying it in many many different problems. 
# tnorm = ISPRSNormal()
from resuneta.src.NormalizeDataset import Normalize
from mxnet.gluon.data.vision import transforms
tnorm = Normalize()

# @@@@@@@@@@@ MODIFY HERE WITH THE ROOT LOCATION WHERE YOU HAVE THE DATA @@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
dataset = ISPRSDataset(root=r'/media/thimabru/ssd/TCC/Resuneta_multasking_amazon/DATASETS/patch_size=256_stride=256_norm_type=1_data_aug=False',mode='train',color=True,mtsk=True,norm=tnorm)
datagen = gluon.data.DataLoader(dataset,batch_size=Nbatch,shuffle=True)
```

```python
from matplotlib import rc
rc("text", usetex=False)
# plt.style.use(['no-latex'])
```

```python
for imgs, labels in datagen:
    with autograd.predict_mode():
        # imgs = gluon.utils.split_and_load(imgs, [ctx])
        preds1, preds2, preds3, preds4 = mynet(imgs.copyto(mx.gpu(0)))
    break

print(imgs.shape)
print(labels.shape)
for idx in range(Nbatch):
    print(f"Nbatch: {idx}")
    print(imgs[idx].shape)
    print(labels[idx].shape)
    view_slice_n_preds(imgs[idx],labels[idx],preds1[idx].asnumpy(),_figsize=(15,8), fig_name='preds.jpg')
```
