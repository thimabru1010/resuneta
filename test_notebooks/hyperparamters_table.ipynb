```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import table # EDIT: see deprecation warnings below
```

```python
data = {'Model': 'unet', 'Dataset': 'amazon',
        'Epochs': 30, 'Optimizer': 'sgd',
        'Learning Rate': 0.1}
    
data_main = {'Hyperparameter': list(data.keys()), 'Value': list(data.values())}

df = pd.DataFrame(data_main, columns=data_main.keys())
# dfStyler = df.style.set_properties(subset=["Unit"],**{'text-align': 'center'})
# dfStyler.set_table_styles([dict(selector='th', props=[('text-align', 'center')])])

# df.style.hide_index()  
# print(dfStyler)
df.align(df, join='center')
df.style.hide_index()  
    
ax = plt.subplot(111, frame_on=False) # no visible frame
ax.xaxis.set_visible(False)  # hide the x axis
ax.yaxis.set_visible(False)  # hide the y axis

table(ax, df, loc='center')  # where df is your data frame

plt.savefig('mytable.png', dpi=200, bbox_inches='tight')
```
